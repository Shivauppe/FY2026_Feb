```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# feature scaling:
#1.Standardization (u =0 , o =1)
#standard scaler(minmax scaler))
#2.Normalization (0-1)
#minmax scaler
#3.unit vector scaling

df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.distplot(df['total_bill'])
plt.show()
```

```python
mean = np.mean(df['total_bill'])
std = np.std(df['total_bill'])
```

```python
std_data = []
for i in list(df['total_bill']):
    z = (i - mean)/std
    std_data.append(z)
```

```python
std_data
```

```python
#feature scaling optional so it does not change distribution
sns.distplot(std_data)
plt.show()
```
